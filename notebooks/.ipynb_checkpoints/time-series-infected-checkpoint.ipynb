```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fname='../model/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv'
```

```python
df = pd.read_csv(fname)
```

```python
df
```

```python
df['Country/Region']
```

```python
usData = df[df['Country/Region'] == 'US']
```

```python
usData
```

```python
progress = usData.loc[:,'1/22/20':]
```

```python
progress
```

```python
y = progress.to_numpy()
```

```python
cummulative = progress.sum()
```

```python
cummulative.plot(marker='o', title='WHO US Time Series - Infected')
```

```python
y = cummulative.to_numpy()
```

```python
y
```

```python
cummulative.index
```

```python
x = np.arange(1, len(y) + 1)
x
```

```python
plt.semilogy(x, y, 'o')
```

```python
m, b = np.polyfit(x, np.log(y), 1)
```

```python
plt.scatter(x, np.log(y))
plt.plot(x, m*x + b, '--')
```

```python
plt.scatter(x, y)
plt.plot(x, np.exp(m*x + b), '--')
plt.xlabel('days')
plt.ylabel('infected')
plt.title('WHO Data - US Covid time series confirmed')
```
